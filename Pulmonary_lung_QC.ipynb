```python
import pandas as pd
from os import system
import os
from Pseudovisium.pseudovisium_generate import generate_pv
```

```python
datasets = pd.read_excel("/content/Datasets_all.xlsx")
datasets_pulmonary = datasets[datasets["Pulmonary Xenium"]==1]
names = datasets_pulmonary["Name"].values
names = [name.split("_")[2] for name in names]

for i in datasets_pulmonary["Onedrive_link"]:
  print(i)
  command="wget {}".format(i)
  system(command)
  filename=i.split("com/")[1]
  command2="unzip /content/{}".format(filename)
  system(command2)
  print(command2)
```

```python
  # Set the directory path
directory = "/content"

# Get a list of all files in the directory
files = os.listdir(directory)
for name in names:
    csv_file=  [file for file in files if file.startswith(name)][0]
    print(csv_file)
    output_path="/content/"
    hexagon_size=50
    pseudovisium_path = generate_pv(csv_file=csv_file,
                                    hexagon_size=hexagon_size,
                                    output_path=output_path,
                                      batch_size=1000000,
                                    technology="Xenium",
                                    max_workers=10,
                                    quality_filter=True,
                                    quality_per_hexagon=True,
                                    quality_per_probe=True,
                                    project_name=name)

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#import /Users/k23030440/Library/CloudStorage/OneDrive-King'sCollegeLondon/spatial_project/pv_qc_2024-04-29_12/probe_quality.csv
probe_quality = pd.read_csv('/content/pv_qc_2024-04-29/probe_quality.csv')
#rename Probe_ID to gene
probe_quality.rename(columns={'Probe_ID':'gene'}, inplace=True)
#/Users/k23030440/Library/CloudStorage/OneDrive-King'sCollegeLondon/spatial_project/pv_qc_2024-04-29_12/sum_stripplot.csv
probe_mean = pd.read_csv('/content/pv_qc_2024-04-29/sum_stripplot.csv')
#/Users/k23030440/Library/CloudStorage/OneDrive-King'sCollegeLondon/spatial_project/pv_qc_2024-04-29_12/morans_i.csv
probe_moransi = pd.read_csv('/content/pv_qc_2024-04-29/morans_i.csv')

#merge all of them on column gene and sample
merged = pd.merge(probe_quality, probe_mean, on=['gene', 'Sample'])
merged = pd.merge(merged, probe_moransi, on=['gene', 'Sample'])
#in merged create a color column which is blue if probe category is not Neg_control and red if it is
merged["color"] = np.where(merged["Probe category"] == "Neg_control", "red", "blue")
#order merged such that red is first then blue
merged = merged.sort_values(by=['color'],ascending=False)
merged.head()
```

```python
#plot scatter between quality and Counts
plt.scatter(merged['log_counts'],merged['Quality'],c=merged['color'],alpha=0.2)
```

```python
plt.scatter(np.log10(merged['Morans_I']),merged['Quality'],c=merged['color'],alpha=0.2)
```

```python
#add a median quality colum, which calculates the median quality for each gene across samples
merged['median_quality'] = merged.groupby('gene')['Quality'].transform('median')
low_quality_genes_df = merged[merged['median_quality']<20]
#remove entries where Probe category is Neg_control
low_quality_genes_df = low_quality_genes_df[low_quality_genes_df['Probe category'] != 'Neg_control']

import seaborn as sns
sns.stripplot(x='gene', y='Quality', data=low_quality_genes_df, alpha=0.5)
plt.xticks(rotation=90)
```

```python
#in merged, find instances when for a probe, Quality is only below 20 in one sample
only_once_low_qual = merged[merged['Quality']<20].groupby('gene').filter(lambda x: len(x) == 1)
only_once_low_qual_genes = only_once_low_qual["gene"].values
#keep only these genes in merged now and make a stripplot
merged_only_once_low_qual_genes=merged[merged["gene"].isin(only_once_low_qual_genes)]
sns.stripplot(x='gene', y='Quality', data=merged_only_once_low_qual_genes, alpha=0.5)
#rotate x tick
plt.xticks(rotation=90)
```
